```r
capacity = c("Production", "MOQ")
wgt_capacity = c(4,1)
m_02=matrix_ahp(capacity, wgt_capacity)
print('Capacity Sub-Criteria: m_02')
round(m_02,3)
```

```r
delivery = c("Lead time", "Logistic Support")
wgt_delivery = c(3,1)
m_03=matrix_ahp(delivery, wgt_delivery)
print('Delivery Sub-Criteria: m_03')
round(m_03,3)
```

```r
risk = c("Reliable", "Currrency Stability")
wgt_risk = c(3,1)
m_05=matrix_ahp(risk, wgt_risk)
print('Risk Sub-Criteria: m_05')
round(m_05,3)
```

```r
sustainability = c("Social", "Environment")
wgt_sustainability = c(2,1)
m_06=matrix_ahp(sustainability, wgt_sustainability)
print('Sustainability Sub-Criteria: m_06')
round(m_06,3)
```

```r
suppliers = c("Southern Monograms", "Hatworx", "Posh Garments", "Repulo Tailors")

# --- 1. Cost ---
# 1.1 Unit Cost
wgt_unitcost = c(6, 3, 7, 10)
m_11 = matrix_ahp(suppliers, wgt_unitcost)

# 1.2 Shipping Cost
wgt_shippingcost = c(7, 9, 4, 2)
m_12 = matrix_ahp(suppliers, wgt_shippingcost)

# --- 2. Capacity ---
# 2.1 Production Capacity
wgt_production = c(4, 2, 10, 8)
m_21 = matrix_ahp(suppliers, wgt_production)

# 2.2 Minimum Order Quantity (MOQ)
wgt_moq = c(10, 6, 5, 10)
m_22 = matrix_ahp(suppliers, wgt_moq)

# --- 3. Delivery ---
# 3.1 Lead Time
wgt_leadtime = c(7, 10, 3, 3)
m_31 = matrix_ahp(suppliers, wgt_leadtime)

# 3.2 Logistics Support
wgt_logisticsupport = c(8, 4, 7, 2)
m_32 = matrix_ahp(suppliers, wgt_logisticsupport)

# --- 4. Payment Terms ---
wgt_paymentterms = c(10, 2, 6, 5)
m_4 = matrix_ahp(suppliers, wgt_paymentterms)

# --- 5. Risk ---
# 5.1 Supply Chain Reliability
wgt_reliable = c(10, 8, 4, 2)
m_51 = matrix_ahp(suppliers, wgt_reliable)

# 5.2 Currency Exchange Rate Stability
wgt_currencyrate = c(10, 10, 5, 4)
m_52 = matrix_ahp(suppliers, wgt_currencyrate)

# --- 6. Sustainability ---
# 6.1 Social Responsibility
wgt_social = c(5, 5, 7, 10)
m_61 = matrix_ahp(suppliers, wgt_social)

# 6.2 Environmental Responsibility
wgt_environment = c(4, 3, 5, 7)
m_62 = matrix_ahp(suppliers, wgt_environment)

# --- Print All Matrices ---
print('1.1 Cost - Unit Cost:')
print(round(m_11, 3))

print('1.2 Cost - Shipping Cost:')
print(round(m_12, 3))

print('2.1 Capacity - Production Capacity:')
print(round(m_21, 3))

print('2.2 Capacity - MOQ:')
print(round(m_22, 3))

print('3.1 Delivery - Lead Time:')
print(round(m_31, 3))

print('3.2 Delivery - Logistics Support:')
print(round(m_32, 3))

print('4. Payment Terms:')
print(round(m_4, 3))

print('5.1 Risk - Reliability:')
print(round(m_51, 3))

print('5.2 Risk - Currency Rate:')
print(round(m_52, 3))

print('6.1 Sustainability - Social:')
print(round(m_61, 3))

print('6.2 Sustainability - Environment:')
print(round(m_62, 3))
```

```r
# map defines the number of subcriteria per criterion - we use it only for the first one (cost)
map <- c(2,2,2,0,2,2) #number of subcriteria for each criteria
# m_0 & m_01 are for criteria & sub-criteria
# m_11, m_12 & m_13 are for alternatives assessed on the first set of subcriteria
# m_2 & m_3 are for alternatives assessed on other criteria
m_ahp<- ahp_s(list(m_0,m_01, m_11, m_12,m_02, m_21, m_22,m_03, m_31, m_32,m_4,m_05, m_51, m_52,m_06, m_61, m_62), map)
m_ahp
```

```r
transforma_tabela(m_ahp)
```

```r
rank <- ranque(m_ahp)
```

```r

# Label criteria and subcriteria
rank$Alternativas <- c("Repulo Tailors","Posh Garments","Southern Monograms", "Hatworx")

# Print final decision matrix with all scores
print("Final AHP Suppliers Ranking:")
rank
```
